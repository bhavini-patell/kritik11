```python
# kritik assignment 4 question 4
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point): # estimate the deriv of function f at initial_point
    x_coords=[initial_point]
    y_coords=[f(initial_point)]

    def f_prime(x):
        return (f(x+0.00001)-f(x))/0.00001
    i=1
    while abs(f_prime(x_coords[i-1]))>0.00001: #checks if deriv if >0.00001, then plot graph cus min is found!
        
        xi=x_coords[i-1] - learning_rate * f_prime(x_coords[i-1]) #new point, resets point every time we iterate the loop

        x_coords.append(xi)
        y_coords.append(f(x_coords[i]))
        i=i+1

    plot_range = np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5, 10000) 
    function_range = [f(i) for i in plot_range]
    plt.plot(plot_range, function_range, label='f(x)') 
    plt.plot(x_coords, y_coords, 'o-', label='Gradient Descent Path')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()
    return round(x_coords[-1], 3), round(y_coords[-1], 3)

def func_1(x):
    return x**2

gradient_descent(func_1, 0.9, 1)
```

```python
def func_2(x):
    return 2*x**4-3*x**3-3*x**2+2*x-1

gradient_descent(func_2, 0.1275, -0.45)
```

```python
def func_3(x):
    if x>0:
        return x**x
    elif x==0:
        return 1
    else:
        return abs(x)**abs(x)

gradient_descent(func_3, 0.875, 1)
```
