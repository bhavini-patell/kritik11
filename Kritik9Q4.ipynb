```python
# Kritik Q4a
import sympy as sp

# Defining symbols
x, y = sp.symbols('x y')

# Defining the function
f = sp.exp(x) * sp.sin(y) + y**3

# Computing partial derivatives
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("Partial derivative of f with respect to x:", df_dx)
print("Partial derivative of f with respect to y:", df_dy)
```

```python
# Kritik Q4b
# Defining the function g
g = x**2 * y + x * y**2

# Computing gradient vector
grad_g = [sp.diff(g, var) for var in (x, y)]

# Evaluating the gradient at the point (1, -1)
grad_g_val = [grad.subs({x: 1, y: -1}) for grad in grad_g]

# Computing magnitude of the gradient vector
magnitude_grad_g_specific = sp.sqrt(sum(comp**2 for comp in grad_g_val))

print("Magnitude of the gradient at (1, -1):", magnitude_grad_g_specific.evalf())
```

```python
# Kritik Q4c
# Defining the function h
h = sp.ln(x**2 + y**2)

# Computing second partial derivatives
d2h_dx2 = sp.diff(h, x, x)
d2h_dy2 = sp.diff(h, y, y)
d2h_dxdy = sp.diff(h, x, y)

print("Second partial derivative of h with respect to x:", d2h_dx2)
print("Second partial derivative of h with respect to y:", d2h_dy2)
print("Second partial derivative of h with respect to x and y:", d2h_dxdy)

# For most smooth functions, the mixed partial derivatives are symmetric:
if d2h_dxdy == d2h_dxdy.subs({x: y, y: x}):
    print("The mixed partial derivatives are symmetric.")
else:
    print("The mixed partial derivatives are not symmetric.")
```

```python
# Kritik Q4d
import numpy as np
import matplotlib.pyplot as plt

# Define the function j using sympy
j = x**3 - 3*x*y + y**3

# Convertng the sympy expression to a numerical function
j_func = sp.lambdify((x, y), j, 'numpy')

# Creating a grid of x and y values and evaluate j on this grid
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

# Ploting the contour
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
# Kritik Q4e

k = x**2 + y**2 #function x^2 + y^2
k_func = sp.lambdify((x, y), k, 'numpy')
Z_k = k_func(X, Y)
plt.contourf(X, Y, Z_k, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = x^2 + y^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
# function f(x, y) = sin(x) * cos(y)
# another one for funsies
f = sp.sin(x) * sp.cos(y)

# Convert the SymPy expression to a numerical function
f_func = sp.lambdify((x, y), f, 'numpy')

# Create a grid of x and y values and evaluate f on this grid
x_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
y_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z_f = f_func(X, Y)

# Plot the contour
plt.contourf(X, Y, Z_f, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $f(x, y) = \sin(x) \cdot \cos(y)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
# Define the funky function g(x, y) = sin(x^2 + y^2) * e^(-(x^2 + y^2))
g = sp.sin(x**2 + y**2) * sp.exp(-(x**2 + y**2))

# Convert the SymPy expression to a numerical function
g_func = sp.lambdify((x, y), g, 'numpy')

# Create a grid of x and y values and evaluate g on this grid
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z_g = g_func(X, Y)

# Plot the contour
plt.contourf(X, Y, Z_g, levels=50, cmap='inferno')
plt.colorbar()
plt.title('Contour plot of $g(x, y) = \sin(x^2 + y^2) \cdot e^{-(x^2 + y^2)}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
# Define the smiley face function
smiley_face = sp.sin(x) * sp.cos(y)

# Convert the SymPy expression to a numerical function
smiley_func = sp.lambdify((x, y), smiley_face, 'numpy')

# Create a grid of x and y values and evaluate the smiley face function on this grid
x_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
y_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z_smiley = smiley_func(X, Y)

# Plot the contour
plt.contourf(X, Y, Z_smiley, levels=50, cmap='coolwarm')
plt.colorbar()
plt.title('Smiley Face Contour Plot')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
